```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data for the bar graph
categories = ["Observed Population", "Repeating 1st", "Repeating 2nd"]
values = [251.67, 317, 220]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['blue', 'green', 'orange'])

# Customization
plt.title("Avg. Days to Destination", fontweight="bold")
plt.ylabel("Days", fontweight="bold")
plt.xticks(fontweight="bold")
plt.yticks(fontweight="bold")

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Data for the "Exit Destination Reported (%)" bar graph
categories = ["Observed Population", "Repeating 1st", "Repeating 2nd"]
values = [30, 34, 45]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['blue', 'green', 'orange'])

# Customization
plt.title("Exit Destination Reported (%)", fontweight="bold")
plt.ylabel("%", fontweight="bold")
plt.xticks(fontweight="bold")
plt.yticks(fontweight="bold")

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Data for the double-bar graph
categories = ["White", "Latinx", "Native", "Black", "Asian"]
population_values = [61, 6, 16, 3, 1]
repeating_values = [73, 7, 20, 2, 1]

# Number of categories
x = np.arange(len(categories))

# Bar width
bar_width = 0.35

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width / 2, population_values, width=bar_width, label="Population", color="blue")
plt.bar(x + bar_width / 2, repeating_values, width=bar_width, label="Repeating", color="green")

# Customization
plt.title("Reported Race/Ethnicity (%)", fontweight="bold")
plt.ylabel("%", fontweight="bold")
plt.xticks(x, categories, fontweight="bold")
plt.yticks(fontweight="bold")
plt.legend(fontsize="medium")

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Data for the triple-bar graph
categories = ["With Ongoing Subsidy", "Without Ongoing Subsidy"]
population_values = [58, 42]
repeating_1st_values = [50, 50]
repeating_2nd_values = [55, 45]

# Number of categories
x = np.arange(len(categories))

# Bar width
bar_width = 0.25

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, population_values, width=bar_width, label="Population", color="blue")
plt.bar(x, repeating_1st_values, width=bar_width, label="Repeating 1st", color="green")
plt.bar(x + bar_width, repeating_2nd_values, width=bar_width, label="Repeating 2nd", color="orange")

# Customization
plt.title("Exit Destination Renting Status", fontweight="bold")
plt.ylabel("%", fontweight="bold")
plt.xticks(x, categories, fontweight="bold")
plt.yticks(fontweight="bold")
plt.legend(fontsize="medium")

# Show plot
plt.tight_layout()
plt.show()
```
